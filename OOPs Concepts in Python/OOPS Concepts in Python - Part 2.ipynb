## Encapsulation

- Encapsulation means Wrapping data(variables) and methods(functions) together in one unit(class) and restricting direct access to data.

- Encapsulation is a basic concept of Object-Oriented Programming (OOP).

- It means binding data and functions together in a class.

- It helps to hide internal details of an object.

- It is also called data hiding.

- Data inside a class is protected from direct access outside the class.

- Access to data is controlled using methods(functions).

- It improves security of data.

- It prevents unauthorized changes to variables.

- Encapsulation is achieved using access modifiers in Python.

- Types of Access Levels:
- Public
- Protected
- Private

- Public variables → can be accessed anywhere.

-  Protected variables → accessed within class and subclasses(_var).

-  Private variables → accessed only inside class (__var).

-  Encapsulation uses getter and setter methods.

-  Getter → used to read data.

- Setter → used to modify data safely.

- It improves code maintainability.

- It makes code easy to manage and update.

- It helps in real-world modeling (like bank account, user data).

- It is an important pillar of OOP (along with inheritance, polymorphism, abstraction).

## Example of Encapsulation

```python
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner          # Public
        self._bank_name = "PyBank"  # Protected
        self.__balance = balance    # Private (name-mangled to _BankAccount__balance)

    # Getter — controlled read access
    def get_balance(self):
        return f"${self.__balance:.2f}"

    # Setter — controlled write access with validation
    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: {self.get_balance()}")
        else:
            print("Deposit amount must be positive.")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. New balance: {self.get_balance()}")
        else:
            print("Invalid or insufficient funds.")
            
# Usage 
acc = BankAccount("Alice", 1000)

acc.deposit(500)        # Deposited $500. New balance: $1500.00
acc.withdraw(200)       # Withdrew $200. New balance: $1300.00
acc.withdraw(9999)      # Invalid or insufficient funds.

print(acc.owner)        # Alice          (public — direct access OK)
print(acc.get_balance()) # $1300.00      (private — accessed via getter)



```

## Abstraction

- Abstraction means Hiding the internal details of how something works and showing only the important features to the user.

- Abstraction is a basic concept of Object-Oriented Programming(OOP).
    
- It means hiding complex implementation details.
    
- Only essential information is shown to the user.
    
- It makes programs simple and easy to use.

- Users don’t need to know how the code works internally.
    
- It reduces program complexity.
    
- It improves code readability.
    
- It helps in focusing on what to do, not how to do.
    
- Abstraction is achieved using abstract classes and methods.
    
- Python provides abstraction using the abc module.
    
- ABC stands for Abstract Base Class.

- Abstract methods are declared using @abstractmethod.
    
- Abstract methods do not have implementation in base class.
    
- Child classes must implement abstract methods.
    
- You cannot create an object of an abstract class directly.
    
- It provides a common structure for subclasses.
    
- It improves security by hiding unnecessary details.
    
- It helps in building large and complex systems.
    
- It supports code reusability and flexibility.

## Example of Abstraction

```python
from abc import ABC, abstractmethod

# Abstract class
class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

# Child class
class Rectangle(Shape):
    def area(self):
        print("Area of Rectangle")

# Child class
class Circle(Shape):
    def area(self):
        print("Area of Circle")

# Creating objects
r = Rectangle()
c = Circle()

r.area()
c.area()
```

## Inheritance 

- Inheritance means One class(child) can use properties and methods of another class(parent) and it helps in reusing code.

- Inheritance is a basic concept of Object-Oriented Programming(OOP).

- It allows one class to inherit properties from another class.

-  The class which gives properties is called Parent Class / Base Class.

-  The class which receives properties is called Child Class / Derived Class.

-  It helps in code reusability.

- It reduces code duplication.

- Child class can use variables and methods of parent class.

- Child class can also add its own new features.

- It improves program structure and organization.

- It supports method overriding(child can change parent method).

- Parent class constructor can be called using super().

- It helps in building real-world relationships.

- Example: Animal -> Dog , Vehicle -> Car etc.

- Types of inheritance in Python: Single, Multiple, Multilevel and Hierarchical and Hybrid.

-  It makes code easy to maintain.

## Example of Inheritance

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Creating object
d = Dog()

d.speak()   # inherited method
d.bark()    # own method
```

## Types of Inheritance in Python

## Single - Level Inheritance

- Single Level Inheritance means A child class inherits from only one parent class and there is one parent - one child relationship,

- One class gives properties -> Parent class

- Another class takes properties -> Child class

- Only one level of inheritance is present

```python
# Generalised Syntax of it:

# class Parent:
    # parent properties

# class Child(Parent):
    # child properties
```

# Example of code

```python
# Parent class
class Animal:
    def speak(self):
        print("Animal makes a sound")

# Child class
class Dog(Animal):
    def bark(self):
        print("Dog barks")

# Create object
d = Dog()

# Call methods
d.speak()   # inherited method
d.bark()    # child method
```

## Multilevel Inheritance

- In multilevel inheritance, a class inherits from a class, which itself inherits from another class.

- It forms a chain of Inheritance(Grandparent -> Parent -> Child).

- One class acts as both parent and child.

- Reuse of code at multiple levels.

## Example of Code

```python
class Animal:
    def speak(self):
        print("Animal makes sound")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Puppy(Dog):
    def weep(self):
        print("Puppy weeps")

p = Puppy()

p.speak()
p.bark()
p.weep()
```

## Multiple Inheritance

- In multiple inheritance, a child class inherits from more than one parent class.

- One child, multiple parents

- Combines features of many classes

- Python supports this

## Example of Code

```python
class Father:
    def skills(self):
        print("Gardening")

class Mother:
    def talent(self):
        print("Cooking")

class Child(Father, Mother):
    def hobby(self):
        print("Painting")

c = Child()

c.skills()
c.talent()
c.hobby()
```

## Hierarchical Inheritance

- In hierarchical inheritance, multiple child classes inherit from one parent class.

- One parent, many children

- Common features are shared

- Child classes have their own methods

```python
class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
    def meow(self):
        print("Cat meows")

d = Dog()
c = Cat()

d.speak()
d.bark()

c.speak()
c.meow()
```

## Hybrid Inheritance

- Hybrid inheritance is a combination of two or more types of inheritance.

- Example: mix of multiple + multilevel.

- Combination of inheritance types.

- More complex.

- Used in advanced programs

## Example of Code

```python
class A:
    def showA(self):
        print("Class A")

class B(A):
    def showB(self):
        print("Class B")

class C(A):
    def showC(self):
        print("Class C")

class D(B, C):
    def showD(self):
        print("Class D")

d = D()

d.showA()
d.showB()
d.showC()
d.showD()
```

## Polymorphism

- Polymorphism means many forms and the same function or method name can be used to perform different tasks depending on the situation.

- Polymorphism is a basic concept of Object-Oriented Programming(OOP).

- The word polymorphism means many forms.

- It allows the same name to behave differently.

- A single function can work in different ways.

- It improves code flexibility.

- It reduces code complexity.

- It makes code more readable and reusable.

- Polymorphism can be achieved using: Method Overloading(limited in Python) and Method Overriding.

- Same method name can be used in different classes.

- The behavior depends on the object calling the method.

- It supports dynamic binding(runtime decision).

- It supports dynamic binding(runtime decision).

- Built-in functions also show polymorphism.

- Example: + operator i.e. 2 + 3 -> addition and "a" + "b" -> string joining.

- It works with functions, operators, and methods.

- Method overriding is the main way in Python.

- Child class can change parent method behavior.

- It helps in real-world problem solving.

```python
# Polymorphism can be achieved by Method Overloading and  Method Overriding.
```

## Method Overriding in Polymorphism

- Method Overriding means A child class provides its own version of a method which is already defined in the parent class.

- Same method name and Same parameters but different implementation in child class.

- When a method is called Python uses the method of the object’s class(child class first).

- Happens in inheritance

- Child class changes parent behavior

- Supports runtime polymorphism

- Python decides method at runtime

- Can use super() to call parent method

- Improves flexibility

- Used in real-world applications.

- Makes code more reusable.

## Examples of Method Overriding

```python
# Basic Overriding

class Animal:
    def sound(self):
        print("Animal makes sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

d = Dog()
d.sound()
```

```python
# Multiple Classes

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        print("Dog barks")

class Cat(Animal):
    def sound(self):
        print("Cat meows")

d = Dog()
c = Cat()

d.sound()
c.sound()
```

```python
# Using super()

class Animal:
    def sound(self):
        print("Animal sound")

class Dog(Animal):
    def sound(self):
        super().sound()   # call parent method
        print("Dog barks")

d = Dog()
d.sound()
```

## Method Overloading in Polymorphism

- Overloading means using the same function or method name nut performing different tasks depending on arguments.

- Overloading = same function name, different parameters and Last function overrides previous ones.

- Not directly supported in Python i.e. Python does not support method overloading directly(like Java/C++)

- Achieved using Default arguments or *args.

- Makes code flexible and reusable.

## Examples of Method Overloading in Polymorphism

```python
# Example using Default Arguments 

class Demo:
    def add(self, a, b, c=0):
        return a + b + c

d = Demo()

print(d.add(2, 3))      # 2 arguments
print(d.add(2, 3, 4))   # 3 arguments
```

```python
# Example using *args

class Demo:
    def add(self, *numbers):
        return sum(numbers)

d = Demo()

print(d.add(2, 3))
print(d.add(2, 3, 4, 5))
```

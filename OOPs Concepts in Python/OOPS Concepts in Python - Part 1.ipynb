## What is OOPS Concepts?

- OOP(Object - oriented Programming System) is a way of writing programs by combining data and behavior together using Classes and Objects.

- It focuses on Real - World Entities.

- Data + Methods are tightly occupied.

- Improves reusability, security, and scalability.

## Functional Programming(FP)

- Programs are written using functions only, and data is passed between them.

- Focuses on Functions.

- Data is mostly seprate.

- Less suited for large, real - world systems.

## Difference between OOPS and Functional Programming.

- Feature 1 : Core Unit:
- OOPS: Object
- FP: Function

- Feature 2: Data Handling:
- OOPS: Inside Objects
- FP: Passed to Functions

- Feature 3: Real - World Modeling:
- OOPS: Easy
- FP: Difficult

- Feature 4: Best for:
- OOPS: Large Applications
- FP: Small Logic - based tasks.

- Feature 5: Example Languages:
- OOPS: Java, C++, Python etc.
- FP: Haskell, Lisp etc.

- Conclusion:
- OOPS binds the data and behavior together, while functional programming keeps them seprate.

## Class Variables vs Instance Variables

- Class Variable:
- Shared by all objects.
- Defined outside methods.

- Instance Variable:
- Unique for each object.
- Defined using self.

```python
# Example of code to demonstrate the difference between the Class Variables and Instance variables.


# CLASS VARIABLES vs INSTANCE VARIABLES in Python


class Car:
    """
    Blueprint for all Car objects.
    """

    
    # Class Variables
    #  Defined Outside __init__
    #  Shared by All objects
    #  Same value for every car (unless changed)
    
    brand = "Toyota"
    wheels = 4
    total_cars = 0       # Will track how many cars are created


   
    # __init__ METHOD (Constructor)
    #  Called automatically when a new object is created
    #  self refers to the specific object being created
   
    def __init__(self, color, model, speed):

        
        # Instance Variables
        #  Defined Inside __init__ using self
        #  Unique to each object
        #  Every car has its OWN color, model, speed
        
        self.color = color      # Unique to this car
        self.model = model      # Unique to this car
        self.speed = speed      # Unique to this car

        # Incrementing class variable each time a car is created
        Car.total_cars += 1     # Updating CLASS variable (use class name)


   
    # Method to display car details
    
    def show_details(self):
        print(f"  Brand  (Class Var)    : {Car.brand}")     # Same for all
        print(f"  Wheels (Class Var)    : {Car.wheels}")    # Same for all
        print(f"  Color  (Instance Var) : {self.color}")    # Unique per car
        print(f"  Model  (Instance Var) : {self.model}")    # Unique per car
        print(f"  Speed  (Instance Var) : {self.speed} kmph") # Unique per car
        print("-" * 42)



# Creating Objects(Instances)
# Each object gets its OWN instance variables
# But Shares the Class variables


print("Creating car1...")
car1 = Car("Red", "Corolla", 180)     # Object 1

print("Creating car2...")
car2 = Car("Blue", "Camry", 200)      # Object 2

print("Creating car3...")
car3 = Car("Black", "Fortuner", 220)  # Object 3

print()



# Displaying the details of each Car.

print(" car1 Details")
car1.show_details()

print("car2 Details")
car2.show_details()

print("car3 Details")
car3.show_details()



# Proving Class Variables are Shared


print(" Class Variable Comparison")
print(f"car1.brand = {car1.brand}")   # Toyota
print(f"car2.brand = {car2.brand}")   # Toyota
print(f"car3.brand = {car3.brand}")   # Toyota
print(" All cars share the SAME brand (Class Variable)")
print()



# PRoving Instance Variables are UNIQUE

print(" Instance Variable Comparison")
print(f"car1.color = {car1.color}")   # Red
print(f"car2.color = {car2.color}")   # Blue
print(f"car3.color = {car3.color}")   # Black
print(" Every car has its OWN color (Instance Variable)")
print()



# Class Variable tracking total objects created


print("Total Cars Created")
print(f"Total Cars = {Car.total_cars}")   # 3
print("Class variable 'total_cars' tracked all objects")

```

## Default Constructor

- A constructor is a special method used to create an object.

- In Python, the constructor is called __init__().

- A default constructor is a constructor that does not take any extra parameters(only self).

- It is used to initialize objects with default values.

- When you create an object, the constructor is automatically called.

- Default constructor helps in setting default values for attributes.

- Now every object will have same default values unless changed.

- It improves code readability.

- It ensures that every object starts with a valid initial state.

- If you do not write any constructor, Python provides a built-in default constructor automatically.

- That built-in constructor does nothing (empty).

- Default constructor is useful when no user input is needed.

- It is different from parameterized constructor, which takes values from user.

- You can still modify values later.

- Default constructor runs only once per object creation.

- It is mainly used in simple programs or initial setup.

## Important Point to be Remembered in OOPS Python:

- We can't have Multiple Constructrors in Python.

```python
# Reason for the above statement:

#1) Python does not support Method Overloading.

# Method Overlading: Method overloading means Having multiple methods with the same name and but with different parameters(number or type).

# Example(Concept – like in Java/C++):
# This is just for understanding (not valid in Python)

# class Demo:
#    def add(self, a, b):
#        return a + b

#    def add(self, a, b, c):
#        return a + b + c
```

```python
#2) Same method name gets overwritten: Python only keeps the latest definition of a method with the same name.

# Example:

# class Demo:
#    def __init__(self):
#       print("First constructor")

#    def __init__(self, name):
#        print("Second constructor")

# Here, the second __init__() replaces the first one.
```

```python
#3) Python is dynamically typed:  Python does not check parameter types or count at compile time.

# Because of this, it cannot decide: Which constructor to call and based on number/type of arguments
```

```python
#4) Only one constructor per class

# In Python, a class can have only one __init__() method.

# That’s why multiple constructors are not allowed.
```

```python
#5) Alternative approach is provided

# Even though multiple constructors are not allowed, Python gives other ways to achieve the same result.
```

## Alternatives to Multiple Constructors

## Default Arguments

- Works like multiple constructors.

```python
# Example:

class Student:
    def __init__(self, name=None, marks=None):
        self.name = name
        self.marks = marks

s1 = Student()
s2 = Student("Abhishek")
s3 = Student("Rahul", 90)
```

## Using *args

```python
class Demo:
    def __init__(self, *args):
        print(args)

d1 = Demo()
d2 = Demo("A")
d3 = Demo("A", 10)
```

## Using Class Methods(Alternative Constructors)

```python
class Student:
    def __init__(self, name):
        self.name = name

    @classmethod
    def from_marks(cls, marks):
        return cls("Marks: " + str(marks))

s1 = Student("Abhishek")
s2 = Student.from_marks(90)
```
